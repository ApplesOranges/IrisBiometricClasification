```python

x = np.zeros((1095, 48, 432))
y = []
count=-1
for i in range(1,220):
  print(i)
  for j in range(1,6):
    if(i<10):
      im = PIL.Image.open(r"/content/drive/MyDrive/IITD_database/Organized/00"+str(i)+"/"+"00"+str(i)+"_"+str(j)+".bmp")
    elif(i<100):
      im = PIL.Image.open(r"/content/drive/MyDrive/IITD_database/Organized/0"+str(i)+"/"+"0"+str(i)+"_"+str(j)+".bmp")
    else:
      im = PIL.Image.open(r"/content/drive/MyDrive/IITD_database/Organized/"+str(i)+"/"+str(i)+"_"+str(j)+".bmp")  
    im = np.asarray(im)
    t, im = cv2.threshold(im, 50, 255, cv2.THRESH_BINARY)
    count+=1
    y.append(str(i))
    for h in range(0, im.shape[0]):
      for w in range(0, im.shape[1]):
        if im[h][w] > 0:
          x[count][h][w]= 1
```

```python
model = tf.keras.models.Sequential([
  tf.keras.layers.Flatten(input_shape=(48, 432)),
  tf.keras.layers.Dense(300,activation='relu'),
  tf.keras.layers.Dense(100,activation='relu'),
  tf.keras.layers.Dense(100,activation='relu'),
  tf.keras.layers.Dense(229,activation='softmax')
])

model.compile(optimizer='adam', 
              loss='sparse_categorical_crossentropy', 
              metrics=['accuracy'])
```

```python
model.fit(x_train, y_train, epochs=30)
```

```python
model.evaluate(x_test, y_test)
```
